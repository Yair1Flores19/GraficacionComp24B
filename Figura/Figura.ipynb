## Rotacion

```python
import numpy as np
x = [1, 2, 2, 1, 1]
y = [1, 1, 2, 2, 1]
plt.plot(x,y)
tetha = np.radians(45)
rotacionX = np.cos(tetha) * np.array(x) - np.sin(tetha) * np.array(y)
rotacionY = np.sin(tetha) * np.array(x) + np.cos(tetha) * np.array(y)
plt.plot(rotacionX, rotacionY)
plt.show()
```
