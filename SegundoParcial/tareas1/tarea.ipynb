```python
import numpy as np
from scipy.linalg import null_space


A1 = np.array([[1, 1, 0],  [1, 1, 0]])


nucleo_A1 = null_space(A1)

print("Núcleo de la primera matriz:")
print(nucleo_A1)

A2 = np.array([[1, 1, 2],[2, 2, 4],[2, 3, 5]])

nucleo_A2 = null_space(A2)

print("\nNúcleo de la segunda matriz:")
print(nucleo_A2)
```
